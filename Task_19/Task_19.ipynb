# Задача 3. Окружность

```python
class Circle:
    def __init__(self, X = 0, Y = 0, R = 1):
        self.X = X
        self.Y = Y
        self.R = R

    def perimeter(self):
        return 2 * 3.14 * self.R

    def square(self):
        return 3.14 * (self.R ** 2)

    def scale(self, k):
        self.R *= k
circle = Circle()
print(circle.perimeter())
print(circle.square())
```

# Задача 4. Отцы, матери и дети

```python
class Parent:

    def __init__(self):
        self.name = 'Mary'
        self.age = 35
        self.child = Child()

    def status(self):
        print(
            'My name is {} i am {} years old.'.format(
                self.name, self.age
            )
        )

    def feed(self):
        print('i Feed my child!')
        child.eat()

    def calmdown(self):
        print('{} is too high. I calm my child down'.format(self.child.calmness))


class Child:

    def __init__(self):
        self.name = 'Timmy'
        self.age = 8
        self.hunger = 0
        self.calmness = 0

    def status(self):
        print(
            'Child HUNGRY = {}({}), Child CALM = {}({})'.format(
                self.hungry(), self.hunger, self.calm(), self.calmness
            )
        )

    def hungry(self):
        if self.hunger > 0:
            return True
        else:
            return False

    def calm(self):
        if self.calmness < 0:
            return False
        else:
            return True

    def play(self):
        print('I am going to play !!!!!')
        self.hunger += 1
        self.calmness -= 1

    def eat(self):
        print('Time to eat')
        self.hunger -= 1
        self.calmness += 1


child = Child()
parent = Parent()

for i in range(1, 11):
    print('\n------DAY {}-------'.format(i))
    child.status()
    if not child.hungry() and child.calm():
        child.play()
    else:
        parent.feed()
    child.status()
```

# Задача 5. Весёлая ферма 

```python
class Potato:
    states = {0: 'Отсутствует', 1: 'Росток', 2: 'Зелёная', 3: 'Зрелая'}

    def __init__(self, index):
        self.index = index
        self.state = 0

    def grow(self):
        if self.state < 3:
            self.state += 1
        self.print_state()

    def is_ripe(self):
        if self.state == 3:
            return True
        return False

    def print_state(self):
        print('Картошка {} сейчас {}'.format(
            self.index, Potato.states[self.state]
        ))


class PotatoGarden:

    def __init__(self, count):
        self.potatoes = [Potato(index) for index in range(1, count + 1)]

    def grow_all(self):
        print('Картошка проростает!')
        for i_potato in self.potatoes:
            i_potato.grow()

    def are_all_ripe(self):
        for i_potato in self.potatoes:
            if not i_potato.is_ripe():
                print('Картошка ещё не созрела!\n')
                break
            else:
                print('Вся картошка созрела! Можно собирать!\n')
```

# Задача 6. Магия

```python
class Fire:
    title = 'Огонь'

    def __add__(self, other):
        if isinstance(other, Air):
            return Lightning()
        elif isinstance(other, Water):
            return Steam()
        elif isinstance(other, Earth):
            return Lava()


class Air:
    title = 'Воздух'

    def __add__(self, other):
        if isinstance(other, Water):
            return Shtorm()
        elif isinstance(other, Fire):
            return Lightning()
        elif isinstance(other, Earth):
            return Dust()


class Water:
    title = 'Вода'

    def __add__(self, other):
        if isinstance(other, Air):
            return Shtorm()
        elif isinstance(other, Fire):
            return Steam()
        elif isinstance(other, Earth):
            return Dirt()


class Earth:
    title = 'Земля'

    def __add__(self, other):
        if isinstance(other, Water):
            return Dirt()
        elif isinstance(other, Fire):
            return Lava()
        elif isinstance(other, Air):
            return Dust()


class Shtorm:
    title = 'Шторм'


class Steam:
    title = 'Пар'


class Dirt:
    title = 'Грязь'


class Lightning:
    title = 'Молния'


class Dust:
    title = 'Пыль'


class Lava:
    title = 'Лава'
```
